```python
import pandas as pd
```

```python
data = pd.read_csv('dataset_Facebook.csv',sep=';',usecols=([1]+list(range(7,19))))
```

```python
types = set(data.Type)
scopes = [[x] for x in types]
scopes.append(types)
metrics = [pd.DataFrame.min, pd.DataFrame.max , pd.DataFrame.mean, pd.DataFrame.median,pd.DataFrame.mode]
pd.set_option('display.max_rows', None)
for metric in metrics:
    frames = []
    print('Metric: ' + metric.__name__)
    for scope in scopes:
        frame = pd.DataFrame(metric(data[data.Type.isin(scope)].loc[:,'Lifetime Post Total Reach':])).rename(index = {0:str(scope)}, columns  = {0:str(scope)})
        if metric != pd.DataFrame.mode:
            frame = frame.transpose()
        frames.append(frame)
    display(pd.concat(frames))
```

```python
## Lifetime Engaged Users - число людей как-то взаимодействовавших с постом
display(data[data['Lifetime Engaged Users'] == (data['Lifetime Engaged Users'].max())])
```

```python
## можно еще посмотреть на отношение таких людей ко всем, кто потенциально видел этот пост
data['c'] = data['Lifetime Engaged Users'] / data['Lifetime Post Total Reach']
sort = data.sort_values('c', ascending = False)
display(sort[sort.c > 0.20])
```
